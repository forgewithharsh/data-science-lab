## Functions with Multiple Arguments

```python
def add(a, b):
    return a + b
print(add(3,5)) # Output: 8
```

```python
def avg(a, b = 10):
    return (a + b) / 2
```

```python
print(avg(4,6))
```

```python
avg(4)
```
