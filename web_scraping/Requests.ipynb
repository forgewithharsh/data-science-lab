## Using requests module for Data Collection

Today we will see how to scrape websites and use requests module to download the raw html of a webpage. In this section we can safely use https://quotes.toscrape.com/ and https://books.toscrape.com/ for scraping demos

```python
import requests
```

```python
a = requests.get("https://quotes.toscrape.com/")
```

```python
for i in range (1, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.index.html")
    with open(f"web_scraping/htmls/page{i}.html", "w") as f:
        f.write(a.text)
        print(f"Downloaded page {i} successfully")
```
