```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [[150,0],
            [170,0],
            [130,1],
            [120,1]]
```

```python
labels = ["apple", "apple", "orange", "orange"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[160, 0], [115, 1]])
```

```python
print(f"The fruit in first list is : {prediction[0]} \nThe fruit in second list is : {prediction[1]}")
```
