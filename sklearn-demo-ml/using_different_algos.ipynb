```python
from sklearn.ensemble import RandomForestClassifier
```

```python
features = [[150,0],
            [170,0],
            [130,1],
            [120,1]]
```

```python
labels = ["apple", "apple", "orange", "orange"]
```

```python
clf = RandomForestClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[150,1]])
```

```python
print(f"The fruit is : {prediction[0]}")
```
